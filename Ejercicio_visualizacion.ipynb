```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/DataSet_Makeover.csv")
df.info()
```

```python
source_counts = df['Source'].value_counts()
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=source_counts.index, y=source_counts.values, palette="husl")
plt.title('Nmero de Artculos por Fuente', fontsize=16)
plt.xlabel('Fuente', fontsize=14)
plt.ylabel('Nmero de Artculos', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

for index, value in enumerate(source_counts.values):
    barplot.text(index, value, str(value), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

```
